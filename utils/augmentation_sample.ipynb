# ImageEnhance plot

```python
import numpy as np
import matplotlib.pyplot as plt
from PIL import Image, ImageOps, ImageEnhance, ImageDraw
```

```python
img_path = '/workspaces/data/ObjectClassify/lemon/train_images/train_0000.jpg'
```

## Origin image

```python
img_ori = Image.open(img_path).convert('RGB')
plt.imshow(img_ori)
plt.show()
```

## AutoContrast

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
img = ImageOps.autocontrast(img)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Brightness

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 1.8
img = ImageEnhance.Brightness(img).enhance(val)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Color

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 1.8
img = ImageEnhance.Color(img).enhance(val)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Contrast

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 1.8
img = ImageEnhance.Contrast(img).enhance(val)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## CutoutConst  
マスクの大きさは論文の結果を見ると画像サイズの1/3~1/2が良さそう[Link](https://www.kumilog.net/entry/numpy-data-augmentation#f-ba95dd14)  
640x640: 213~320

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 40
w, h = img.size
x0 = np.random.uniform(0, w)
y0 = np.random.uniform(0, h)
x0 = int(max(0, x0 - val / 2.))
y0 = int(max(0, y0 - val / 2.))
x1 = int(min(w, x0 + val))
y1 = int(min(h, y0 + val))
xy = (x0, y0, x1, y1)
# gray
color = (127, 127, 127)
img = img.copy()
ImageDraw.Draw(img).rectangle(xy, color)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Equalize

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
img = ImageOps.equalize(img)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Invert

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
img = ImageOps.invert(img)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Posterize

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 1 # 1~9 (0: 真っ暗)
img = ImageOps.posterize(img, val)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Rotate

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 90 # 1~9 (0: 真っ暗)
img_90 = img.rotate(val)
img_r90 = img.rotate(-val)

plt.subplot(1, 3, 1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1, 3, 2)
plt.title('transformed')
plt.imshow(img_90)
plt.subplot(1, 3, 3)
plt.title('transformed')
plt.imshow(img_r90)
plt.show()
```

## Sharpness

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 10
img = ImageEnhance.Sharpness(img).enhance(val)

plt.subplot(1,2,1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1,2,2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## ShearX

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 0.3
img = img.transform(img.size, Image.AFFINE, (1, val, 0, 0, 1, 0), Image.BILINEAR)

plt.subplot(1, 2, 1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1, 2, 2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## ShearY

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 0.3
img = img.transform(img.size, Image.AFFINE, (1, 0, 0, val, 1, 0), Image.BILINEAR)

plt.subplot(1, 2, 1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1, 2, 2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## Solarize

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 200
img = ImageOps.solarize(img, 256 - val)

plt.subplot(1, 2, 1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1, 2, 2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```

## SolarizeAdd

```python
img_ori = Image.open(img_path).convert('RGB')
img = Image.open(img_path).convert('RGB')
val = 200
threshold=128

img_np = np.array(img).astype(np.int)
img_np = img_np + val
img_np = np.clip(img_np, 0, 255)
img_np = img_np.astype(np.uint8)
img = Image.fromarray(img_np)

img = ImageOps.solarize(img, threshold)

plt.subplot(1, 2, 1)
plt.title('origin')
plt.imshow(img_ori)
plt.subplot(1, 2, 2)
plt.title('transformed')
plt.imshow(img)
plt.show()
```
